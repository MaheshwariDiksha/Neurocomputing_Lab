```python
from sklearn.datasets import make_regression
```

```python
N = 100
X, t = make_regression(n_samples=N, n_features=1, noise=15.0)
```

```python
print(X.shape)
print(t.shape)

plt.figure(figsize=(10, 5))
plt.scatter(X, t)
plt.xlabel("x")
plt.ylabel("t")
plt.show()
```

```python
w = 0
b = 0

eta = 0.1

for epoch in range(100):
    dw = 0
    db = 0.0
    
    for i in range(N):
        # Prediction
        y = w * X[i] + b
        
        # LMS
        dw += (t[i] - y) * X[i]
        db += (t[i] - y)
        
    # Parameter updates
    w += eta * dw / N
    b += eta * db / N
    
print(w, b)
```

```python
plt.figure(figsize=(10, 5))

plt.scatter(X, t)

x_axis = [X.min(), X.max()]
plt.plot(x_axis, w*x_axis + b)

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
y = w * X + b

plt.figure(figsize=(10, 5))

plt.scatter(t, y)

x_axis = [t.min(), t.max()]
plt.plot(x_axis, x_axis)

plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
w = 0
b = 0

eta = 0.1

losses = []
for epoch in range(100):
    dw = 0
    db = 0.0
    mse = 0.0
    
    for i in range(N):
        # Prediction
        y = w * X[i] + b
        
        # LMS
        dw += (t[i] - y) * X[i]
        db += (t[i] - y)
        
        # mse
        mse += (t[i] - y)**2
    
    # Parameter updates
    w += eta*dw/N
    b += eta*db/N
    losses.append(mse/N)
    
print('mse:', mse/N)
    
plt.figure(figsize=(10, 5))
plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("mse")
plt.show()
```

```python
from sklearn.datasets import make_regression
```

```python
N = 100
X, t = make_regression(n_samples=N, n_features=1, noise=0.0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Just to avoid the annoying warnings, please ignore
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
 
def create_dataset(N, noise):
    "Creates a dataset of N points generated from x*sin(x) plus some noise."
    
    x = np.linspace(0, 10, 300)
    rng = np.random.default_rng()
    rng.shuffle(x)
    x = np.sort(x[:N])
    t = x * np.sin(x) + noise*rng.uniform(-1.0, 1.0, N)
    
    return x, t

N = 16
X, t = create_dataset(N, noise=0.2)
```

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 6))
plt.plot(x, x*np.sin(x), label="Ground truth")
plt.scatter(X, t)
plt.legend()
plt.show()
```

```python
deg = 10

# Polynomial regression
w = np.polyfit(X, t, deg)

# Inference on the training set
y = np.polyval(w, x)

plt.figure(figsize=(10, 6))
plt.plot(x, x*np.sin(x), label="Ground truth")
plt.plot(x, y, label="Polynome")
plt.scatter(X, t)
plt.ylim((-10., 10.))
plt.legend()
plt.show()
```

```python
training_mse = []

degrees = range(1, 21)

for deg in degrees:
    
    w = np.polyfit(X, t, deg)
    y = np.polyval(w, X)
    
    mse = np.mean((t-y)**2)
    
    training_mse.append(mse)
    
    print("Degree", deg, ": training error", mse)
    
plt.figure(figsize=(10, 6))
plt.plot(degrees, training_mse)
plt.xlabel("Order of the polynomial")
plt.ylabel("Training mse")
plt.show()
```

```python
N_train = int(0.7*N)
X_train, t_train = X[:N_train], t[:N_train]
X_test, t_test = X[N_train:], t[N_train:]
```

```python
print(X_train)
print(t_train)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, t_train, t_test = train_test_split(X, t, test_size=0.3)

degrees = range(1, 21)

test_mse = []

for deg in degrees:
    # Train on the training set
    w = np.polyfit(X_train, t_train, deg)
    
    # Test on the test set
    y_test = np.polyval(w, X_test)
    mse = np.mean((t_test-y_test)**2)
    test_mse.append(mse)
    
    print("Degree", deg, ": empirical error", mse)
print(np.amin(test_mse))    
plt.figure(figsize=(10, 6))
plt.plot(degrees, test_mse)
plt.xlabel("Order of the polynomial")
plt.ylabel("Test mse")
plt.show()
```

```python
from sklearn.model_selection import KFold

k = 4
kf = KFold(n_splits=k, shuffle=True)
```

```python
for train_index, test_index in kf.split(X, t):
    print("Train:", train_index)
    print("Test:", test_index)
    print('-------')
```

```python
k = N
kf = KFold(n_splits=k, shuffle=True)

degrees = range(1, 21)

test_mse = []

for train_index, test_index in kf.split(X, t):
    
    split_mse = []
    
    for deg in degrees:
        
        w = np.polyfit(X[train_index], t[train_index], deg)
        y = np.polyval(w, X[test_index])
        
        mse = np.mean((t[test_index] - y)**2)
        split_mse.append(mse)
    
    test_mse.append(split_mse)
print(test_mse)        
test_mse = np.mean(test_mse, axis=0)
print(np.amin(test_mse))

for deg, mse in zip(degrees, test_mse):
    print(deg, mse)

plt.figure(figsize=(10, 6))
plt.plot(degrees, test_mse)
plt.xlabel("Degree of the polynome")
plt.ylabel("k-fold cross-validated mse")
plt.show()
```
